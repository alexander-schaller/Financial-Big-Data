```python
import pandas as pd
import datetime as dt
import seaborn as sns
import os
```

```python
def clean_timestamp(data):
    data["datetime"] = pd.TimedeltaIndex(data["xltime"], unit='d') + dt.datetime(1899,12,30)
    data["datetime"] = data["datetime"].dt.tz_localize('UTC').dt.tz_convert('America/New_York')
    data = data.set_index("datetime").drop(columns=["xltime"])
    return data
```

```python

def save_clean_timestamp(data, name):
    data = clean_timestamp(data)
    data.to_parquet('../Data/clean/' + name + '.parquet')
```

```python

SPY = pd.read_parquet('../Data/raw/2009-12-08-SPY.P-trade.parquet')
save_clean_timestamp(SPY, "2009-12-08-SPY.P-trade")
```
