```python
import pandas as pd
```

```python
data = pd.read_csv('../Data/us_equities.csv')

data.drop(['Unnamed: 0'], axis=1, inplace=True)
data["Missing Values"] = data.isnull().sum(axis=1)
display(data.head())
data["Missing Values"].plot()
```

```python
gspc_null = data[data["^GSPC"].isnull()]
display(gspc_null)
weird_assets = gspc_null.notnull().sum() > 1
display(weird_assets.drop(weird_assets == False))
```
